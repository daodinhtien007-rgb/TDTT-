```python
#W8A1
class Cylinder:
    def __init__(self,bk,ccao):
        self.bk=bk
        self.ccao=ccao
    def dien_tich(self):
        s=2*3.14*(self.bk**2)+2*3.14*self.bk*self.ccao
        return round(s,2)
    def the_tich(self):
        V=3.14*(self.bk**2)*self.ccao
        return round(V,2)
r,h=map(float,input().split())
Cylinder1=Cylinder(r,h)
print(Cylinder1.dien_tich())
print(Cylinder1.the_tich())
```

```python
#W8A2
class Date:
    def __init__(self,ngay,thang,nam):
        self.ngay=ngay
        self.thang=thang
        self.nam=nam
    def nam_nhuan(self):
        return self.nam%400==0 or (self.nam%4==0 or self.nam%100!=0)
    def so_ngay_trong_thang(self):
        if self.thang in [1,3,5,7,8,10,12]:
            return 31
        elif self.thang in [2,4,6,9,11]:
            return 30
        elif self.thang==2:
            return 29 if self.nam_nhuan() else 28
        else:
            return 0
    def hop_le(self):
        if self.nam<1 or self.thang<1 or self.thang>12:
            return False
        if self.ngay<1 or self.ngay>self.so_ngay_trong_thang():
            return False
        return True
    def ngay_ke_tiep(self):
        if not self.hop_le():
            return None
        ngay=self.ngay+1
        thang=self.thang
        nam=self.nam
        if ngay>self.so_ngay_trong_thang():
            ngay=1
            thang+=1
            if thang>12:
                thang=1
                nam+=1
        return Date(ngay,thang,nam)

n=input().strip()
try:
    a,b,c=map(int,n.split("/"))
    ngay=Date(a,b,c)
    if not ngay.hop_le():
        print("INVALID")
    else:
        next_day=ngay.ngay_ke_tiep()
        print(f"{next_day.ngay}/{next_day.thang}/{next_day.nam}")
except:
    print("INVALID")
```

```python
#W8A3
import math
class Point:
    def __init__(self,x,y):
        self.x=x
        self.y=y
    def goc_toa_do(self):
        return self.x==0 and self.y==0
    def tren_truc_tung(self):
        return self.x==0 and self.y!=0
    def tren_truc_hoanh(self):
        return self.x!=0 and self.y==0
    def khoang_cach(self,goc):
        return math.sqrt((self.x-goc.x)**2+(self.y-goc.y)**2)
x,y=map(float,input().split())
A=Point(x,y)
O=Point(0,0)
if A.goc_toa_do():
    print("A tren goc toa do")
else:
    print("A khong la tren goc toa do")
if A.tren_truc_tung():
    print("A nam tren truc tung")
else:
    print("A khong nam tren truc tung")
if A.tren_truc_hoanh():
    print("A nam tren truc hoanh")
else:
    print("A khong nam tren truc hoanh")

kc=A.khoang_cach(O)
print(f"Khoang cach tu A den goc toa do la: {kc:.2f}")
    
```

```python
#W8A6
class Rectangle:
    def __init__(self,w,h):
        self.w=w
        self.h=h
    def area(self):
        return self.w*self.h
    def perimeter(self):
        return 2*(self.w+self.h)
    def scale(self,k):
        self.w=self.w*k
        self.h=self.h*k
w,h,k=map(float,input().split())
a=Rectangle(w,h)
a.scale(k)
area=round(a.area(),2)
perimeter=round(a.perimeter(),2)
print(f"{area:.2f} {perimeter:.2f}")
```

```python
#W8A7
import math
class Fraction:
    def __init__(self,num,den):
        if den<0:
            num=-num
            den=-den
        if den!=0:
            k=math.gcd(num,den)
            self.num=num//k
            self.den=den//k
    def add(self,other):
        num=self.num*other.den+other.num*self.den
        den=self.den*other.den
        return Fraction(num,den)
    def sub(self,other):
        num=self.num*other.den-other.num*self.den
        den=self.den*other.den
        return Fraction(num,den)
    def mul(self,other):
        num=self.num*other.num
        den=self.den*other.den
        return Fraction(num,den)
    def div(self,other):
        num=self.num*other.den
        den=self.den*other.num
        return Fraction(num,den)
    def __str__(self):
        return f"{self.num}/{self.den}"
    
a,b,op,c,d=input().split()
m=Fraction(int(a),int(b))
n=Fraction(int(c),int(d))
if op=="+":
   kq= m.add(n)
if op=="-":
    kq=m.sub(n)
if op=="*":
    kq=m.mul(n)
if op=="/":
    kq=m.div(n)
print(kq)

    
```

```python
#W8A9
class Student:
    def __init__(self,name):
        self.name=name
        self.score={}
    def add_score(self,subject,score):
        self.score[subject]=score
    def avg(self):
        sum=0
        a=0
        for i in self.score:
            sum+=self.score[i]
            a+=1
        return sum/a
    def rank(self):
        if self.avg() >=8:
            return "Excellent"
        elif self.avg() >=6.5:
            return "Good"
        elif self.avg() >=5:
            return "Average"
        else:
            return "Poor"
    
name=input()
n=int(input())
Student=Student(name)
for i in range(n):
    a,b=input().split()
    Student.add_score(a,float(b))
print(f"{name} {Student.avg():.2f} {Student.rank()}")

```

```python
#W8A12
class  ComplexNumber:
    def __init__(self,double_initRealPart, double_initImaginaryPart):
        self.RealPart=double_initRealPart
        self.ImaginaryPart=double_initImaginaryPart
    def ComplexNumber_multiply(self,other):
        double_initRealPart=self.RealPart*other.RealPart-self.ImaginaryPart*other.ImaginaryPart
        double_initImaginaryPart=self.RealPart*other.ImaginaryPart+self.ImaginaryPart*other.RealPart
        return ComplexNumber(double_initRealPart,double_initImaginaryPart)
    def __str__(self):
        return f"{self.RealPart} {self.ImaginaryPart}"
a,b=map(int,input().split())
c,d=map(int,input().split())
n=ComplexNumber(a,b)
k=ComplexNumber(c,d)
m=n.ComplexNumber_multiply(k)
print(m)
```

```python
#W8A13
import math
class Triangle:
    def __init__(self,initEdge1,initEdge2,initEdge3):
        self.edge1=initEdge1
        self.edge2=initEdge2
        self.edge3=initEdge3
    def double_getArea(self):
        a,b,c=self.edge1,self.edge2,self.edge3
        if a+b<=c or a+c<=b or b+c<=a:
            return -1
        p=(a+b+c)/2
        Area=math.sqrt(p*(p-a)*(p-b)*(p-c))
        return Area
a,b,c=map(int,input().split())
Triangle=Triangle(a,b,c)
Area=Triangle.double_getArea()
if Area==-1:
    print("invalid")
else:
    print(f"{Area:.2f}")
```
