```python
#W7A1
def binary_search(arr,left,right,target):
    while left<=right:
        mid=(left+right)//2
        if arr[mid]==target:
            return mid
        elif arr[mid]<target:
            left=mid+1
        else :
            right=mid-1
    return -1
arr=list(map(int,input().split()))
target=int(input())
print(binary_search(arr,0,len(arr)-1,target))
```

```python
#W7A2
def count_occurrences(arr,x) ->int:
    count=0
    for i in arr:
        if i==x:
            count+=1
    return count
arr=list(map(int,input().split()))
x=int(input())
print(count_occurrences(arr,x))
```

```python
#W7A3
def selection_sort(arr):
    for i in range(len(arr)):
        min_idx=i
        for j in range(i+1,len(arr)):
            if arr[j]<arr[min_idx]:
                min_idx=j
        arr[i],arr[min_idx]=arr[min_idx],arr[i]
    for k in arr:
        print(k,end=" ")
arr=list(map(int,input().split()))
selection_sort(arr)
```

```python
#W7A3
def bubble_sort(arr):
    a=len(arr)
    for i in range(a):
        swapped=False
        for j in range(a-i-1):
            if arr[j]>arr[j+1]:
                arr[j],arr[j+1]=arr[j+1],arr[j]
                swapped=True
        if not swapped:
            break
    for k in arr:
        print(k,end=" ")
arr=list(map(int,input().split()))
bubble_sort(arr)
```
