```python
#W11A1
def reverse(path):
    try:
        f=open(path,'r')
        n=f.read().strip()
        f.close()
        return int(n[::-1])
    except:
        return -1
print(reverse("test.txt"))
```

```python
#W11A2
def canWinNim(path):
    file = open(path, "r")
    a= int(file.readline().strip()) 
    file.close()
    if a%4==0:
        return False
    return True
print(canWinNim("test.txt"))
```

```python
#W11A4
def areAnagrams(path1: str, path2: str):
    f1=open(path1,'r')
    s1=f1.read().lower()
    f2=open(path2,'r')
    s2=f2.read().lower()
    a=[]
    b=[]
    f1.close()
    f2.close()
    for i in s1:
        if i.isalpha():
            a.append(i)
    for j in s2:
        if j.isalpha():
            b.append(j)
    if sorted(a)==sorted(b):
        return True
    return False
print(areAnagrams("test.txt","test2.txt"))
```

```python
#W11A5
def count(path,character):
    f=open(path,"r")
    a=f.read()
    f.close()
    print(a.count(character))
count("test.txt","e")
```

```python
#W11A6
def countCharacter(path,letter):
    f=open(path,'r')
    a=f.read().lower()
    f.close()
    return a.count(letter)
print(countCharacter("test.txt",'e'))
```

```python
#W11A7
def open_file(path):
    try:
        f=open(path,'r')
        f.close()
        return 'YES'
    except:
        return 'NO'
print(open_file(input()))
```

```python
#W11A8
def fib(n):
    f=[""]*(n+1)
    f[0]="a"
    f[1]="b"
    for i in range(2,n+1):
        f[i]=f[i-1]+f[i-2]
    file=open("out.txt",'w')
    for j in range(n+1):
        file.write(f"f({j})={f[j]}\n")
fib(20)
```
