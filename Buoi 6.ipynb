```python
#W6A12
n=list(input().split())
so_lan_max=0
so_max=n[0]
for i in n:
    dem=n.count(i)
    if dem>so_lan_max:
        so_lan_max=dem
        so_max=i
    elif dem==so_lan_max and so_max>i:
        so_max=i
print(so_max)
```

```python
#W6A14
def xoa_ptu_trung(a):
    b=[]
    for i in a:
        if i not in b:
            b.append(i)
    return b
m=list(map(int,input().split()))
n=list(map(int,input().split()))
c=[]
for i in xoa_ptu_trung(m):
    for j in xoa_ptu_trung(n):
        if i==j:
            c.append(i)
print(c)


```
