```python
#W9A1


def merge(a,b):
    res=[]
    i=j=0
    while i<len(a) and j <len(b):
        if a[i]<=b[i]:
            res.append(a[i])
        else:
            res.append(b[i])
        i+=1
    while i<len(a):
        res.append(a[i])
        i+=1
    while i<len(b):
        res.append(b[i])
        i+=1
    return res

a=list(map(int,input().split()))
b=list(map(int,input().split()))
print(merge(a,b))
```

```python
#W9A4
def sosanh(a,b):
    c=a.upper()
    for i in range(len(a)-len(b)+1):
        if c[i:i+3]==b:
            return "YES"
    return "NO"
n=int(input())
d=[]
for i in range(n):
    a=input()
    b=input()
    d.append([a,b])
for i in d:
    a=i[0]
    b=i[1]
    print(sosanh(a,b))
```

```python
#W9A5
def lst(lst1,lst2):
    a=len(lst1)
    b=len(lst2)
    if a>b:
        return "NO"
    else:
        for i in range(b-a+1):
            if lst2[i:i+a]==lst1:
                return "YES"
        return "NO"
lst1=list(map(int,input().split()))
lst2=list(map(int,input().split()))
print(lst(lst1,lst2))
```

```python
#W9A7
def canPlaceFlowers(a,k):
    for i in range(len(a)):
        if a[i]==0 and len(a)!=1:
            if (i==0 and a[i+1]==0) or (i==len(a)-1 and a[i-1]==0) or (a[i-1]==0 and a[i+1]==0):
                a[i]=1
                k=k-1
                
        elif len(a)==1:
            if a[i]==0:
                a[i]=1
                k=k-1
        if k==0:
            return True
    return False
a=list(map(int,input().split()))
k=int(input())
print(canPlaceFlowers(a,k))
```

```python
#W9A8
def is_path_crossing(moves):
    a=[[0,0]]
    b=list(moves)
    for i in range(len(b)):
        if b[i]=="R":
            x=a[i][0]+1
            a.append([x,a[i][1]])
        elif b[i]=="L":
            x=a[i][0]-1
            a.append([x,a[i][1]])
        elif b[i]=="U":
            y=a[i][1]+1
            a.append([a[i][0],y])
        elif b[i]=="D":
            y=a[i][1]-1
            a.append([a[i][0],y])
    c=[]
    for i in a:
        if i not in c:
            c.append(i)
    if len(c)==len(a):
        return False
    return True
            
moves=input()
is_path_crossing(moves)
```
