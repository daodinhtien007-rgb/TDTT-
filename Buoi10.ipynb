```python
#W10A1
def divide(a,b):
    try:
        m=a/b
        print(f"{m:.2f}")
    except ZeroDivisionError:
        print("Loi chia khong")

a,b=map(int,input().split())
divide(a,b)
```

```python
#W10A3
def tong_bp(n):
    a=0
    while n!=0:
        b=n%10
        a+=b**2
        n=n//10
    return a
def so_hp(n):
    a=set()
    while n!=1:
        if n in a:
            return False
        a.add(n)
        n=tong_bp(n)
    return True
try:
    n=int(input())
    if so_hp(n):
        print("YES")
    else:
        print("NO")
except ValueError as e:
    print(e)
```

```python
#W10A6
def can_bac_hai(x):
    return float(x)**(1/2)
try:
    x=float(input())
    if x<0:
        raise ValueError
    print(f"{can_bac_hai(x):.2f}")
except ValueError:
    print("So am")
```

```python
#W10A10
def arr(n):
    s=0
    if len(n)!=len(set(n)):
        raise ValueError
    for i in n:
        s+=i
    return s
    
try:
    n=list(map(int,input().split()))
    N=n[0]
    a=n[1:]
    print(arr(a))
except ValueError:
    print("Mang khong hop le")
```
