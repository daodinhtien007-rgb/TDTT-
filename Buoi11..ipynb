```python
#W11A1
def reverse(path):
    try:
        f=open(path,'r')
        n=f.read().strip()
        f.close()
        return int(n[::-1])
    except:
        return -1
print(reverse("test.txt"))
```

```python
#11A2
def canWinNim(path):
    file = open(path, "r")
    a= int(file.readline().strip()) 
    print(a)
    file.close()
    if a%4==0:
        return False
    return True
print(canWinNim("test.txt"))
```
