```python
#W4A30
n = input("Nhap chuoi: ")
ki_tu_hoa=ki_tu_thuong=ki_tu_so=0
for i in n:
    if i.isupper():
        ki_tu_hoa+=1
    elif i.islower():
        ki_tu_thuong+=1
    elif i.isdigit():
        ki_tu_so+=1
print(f"{ki_tu_hoa} hoa")
print(f"{ki_tu_thuong} thuong")
print(f"{ki_tu_so} so")
```

```python
#W4A32
n=input("Nhap mat khau: ")
tong_1=tong_2=tong_3=tong_4=0
for i in n:
    if i.isupper():
        tong_1+=1
    elif i.islower():
        tong_2+=1
    elif i.isdigit():
        tong_3+=1
    else:
        tong_4+=1
if tong_1>=1 and tong_2>=1 and tong_3>=1 and tong_4>=1 and (tong_1 + tong_2 + tong_3 + tong_4 )>6:
    print("Mat khau manh")
else:
    print("Ban nen dat mat khau manh hon")
```
